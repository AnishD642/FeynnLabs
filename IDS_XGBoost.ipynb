```python
from sklearn.preprocessing import LabelEncoder
labelen = LabelEncoder()
df['protocol_type'] = labelen.fit_transform(df['protocol_type']) 
df['service'] = labelen.fit_transform(df['service'])
df['flag'] = labelen.fit_transform(df['flag'])
df['class'] = labelen.fit_transform(df['class'])
```

```python
df.head()
```

```python
X=df.iloc[:,0:41]
y=df.iloc[:,41:42]
```

```python
X.head()
```

```python
X.shape
```

```python
y.head()
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=10)
```

```python
from xgboost import XGBClassifier
model = XGBClassifier(n_estimators=200, learning_rate=0.6, max_depth=5, random_state=10)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix
sns.set(rc={'figure.figsize':(4,4)})
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm,annot=True)
```
